**Importing necessary libraries**

```python
#importing all libraries
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

**Data Colllection**

```python
#Load the apps dataset
apps_dada=pd.read_csv('apps.csv')
```

```python
##check the head of dataset
print("Display the apps data to get overview:")
apps_dada.head(10)
```

```python
##check the bottom of dataset
apps_dada.tail(10)
```

**Data Preparation**

```python
#shape of the our dataset
print("The shape of our dataset: ",apps_dada.shape)
```

In this data, the data contains 9659 rows and 14 columns.



```python
#check data type
apps_dada.dtypes
```

In that we can see that wrong data types and remove unwanted characters from the 'Installs' and 'Price' variables

Data cleaning

```python
#firstly remove unwanted characters from Installs and prices coloumns
#installs column
#Remove the Unwanted characters

char_to_remove = ['+',',','$']
col_contain_char = ['Installs','Price']

for col in col_contain_char:
  for i in char_to_remove:
    apps_dada[col] = apps_dada[col].apply(lambda x:x.replace(i,''))

apps_dada.head()
```

Correct data type

```python
#convert data type in installs column
apps_dada['Installs'] = apps_dada['Installs'].astype('float')
#convert data type in price column
apps_dada['Price'] = apps_dada['Price'].astype('float')
```

```python
apps_dada.dtypes
```

**Category Exploration**

1.How many different App Categories are there in the dataset?

```python
#calculate different app categories
categories=apps_dada.Category.value_counts()
print('App Catergories        count')
print(' ')
print(categories)
print('Total number of categories :',len(categories))
```

In that we can see that, 

- The highest number of apps in avaible on google play store is Family and Game category.
- The Lowest number of apps in avaible on google play store is Beauty category.
- Total number of apps category avaible is 33 different category

2.What is the average rating of apps in each category?



```python
#Average rating apps
Average_rating=apps_dada.groupby('Category')['Rating'].mean().sort_values(ascending=False)
print("The average rating of apps in each category:",Average_rating)
```

3.What is the average number of reviews for Apps in each Category?



```python
#Average review 
Average_review = apps_dada.groupby('Category')['Reviews'].mean().sort_values(ascending=False)
print("The average number of reviews for Apps in each Category:",Average_review)
```

```python
plt.figure(figsize=(10,8))
Average_review.plot(kind='bar',color='royalblue')
plt.title('Average Reviews of Apps across Categories')
plt.xlabel('Categories')
plt.ylabel('Average Number of Reviews')
plt.xticks(rotation=90)
plt.show()
```

4.Which are the top 5 Categories with highest number of installs?

```python
#top 5 categories
top_install_by_cat = apps_dada.groupby('Category')['Installs'].sum().sort_values(ascending=False)
print('Top 5 Categories with highest number of installs :\n',top_install_by_cat.head(5))
```

5.How apps are distributed among different categories in the dataset?

```python
#display different apps categories
plt.figure(figsize=(12,8))
sns.histplot(data=apps_dada,x='Category',kde=False,edgecolor='black',color='royalblue')
plt.xlabel('Categories')
plt.ylabel('Number of Apps')
plt.title('Distribution of Apps across Categories')
plt.xticks(rotation=90)
plt.show()
```

6.What is the distribution of app type(Free or Paid) within each category?



```python
app_dist = apps_dada.pivot_table(index='Category',columns='Type',aggfunc='size',fill_value=0)
plt.figure(figsize=(10,8))
app_dist.plot(kind='barh',stacked=True)
plt.xlabel('Category')
plt.ylabel('Number of Apps')
plt.title('Distribution of App types across Categories')
plt.legend(title='App Types')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```

In that we can see that, the majority of apps belong to Free Type across all Categories.


7.What is the distribution of App prices within each category?



```python
#Displayed distribution of App prices within each category
plt.figure(figsize=(10,8))
sns.barplot(data=apps_dada,x='Category',y='Price',color='royalblue')
plt.title('Distribution of App prices within categories')
plt.xlabel('Category')
plt.ylabel('App Price($)')
plt.xticks(rotation=90)
plt.show()
```

In this chart, Finance and lifestle apps highest price apps

**Sentiment Analysis**

In that we will load review dataset and Assess user sentiments through reviews

```python
#load reviews datasets
user_reviews=pd.read_csv("user_reviews.csv")
user_reviews.head(10)
```

```python
#shape of the our dataset
print("The shape of our dataset: ",user_reviews.shape)
```

```python
#now combine both datasets
combine_data = pd.merge(apps_dada,user_reviews,on='App')
combine_data.head()
     
```

Distribution of sentiment polarity

```python
merge_df=combine_data.dropna(subset=['Sentiment','Translated_Review'])

#Analyse the results
#User review sentiment polarity for paid vs. free apps
sns.set_style('ticks')
fig,ax = plt.subplots()
fig.set_size_inches(11,8)

ax = sns.boxplot(x='Sentiment_Polarity',y='Type',data=merge_df)
ax.set_title('Distribution of sentiment polarity')
```

Sentiment Polarity across Categories

```python
plt.figure(figsize=(10,8))
sns.barplot(x='Category',y='Sentiment_Polarity',data=merge_df,color='royalblue')
plt.xlabel('Categories')
plt.ylabel('Sentiment Polarity')
plt.title('Sentiment Polarity across Categories')
plt.xticks(rotation=90)
plt.show()
```
